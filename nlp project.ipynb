```python
model = gensim.models.Word2Vec([words1, words2], min_count=1)
```

```python
model.wv.vocab
```

```python
model.build_vocab(['Data'], update=True)
model.train(['Data'], total_examples=1, epochs=1)
```

```python
# Create word embeddings for each document
vec1 = np.mean([model.wv[word] for word in words1], axis=0)
vec2 = np.mean([model.wv[word] for word in words2], axis=0)

```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity([vec1], [vec2])[0][0]
print(f'Similarity between the two documents: {similarity:.3f}')
```
